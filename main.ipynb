#### Оскільки в нашому наборі даних міститься інформація про вік від 15 до 18 років, то скоріш за все аномалій не буде

```python
# Функція для виявлення аномалій і видалення їх з DataFrame
def remove_anomalies(df):
    Q1 = df['Age'].quantile(0.25)
    Q3 = df['Age'].quantile(0.75)
    IQR = Q3 - Q1
        
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
        
    # Видалення аномалій
    df = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]
    
    return df

# Видалення аномалій з DataFrame
cleaned_data = remove_anomalies(data)

print("Очищений DataFrame без аномалій:")
print(cleaned_data)
```

#### Аналіз впливу додаткових занять на академічні успіхи

```python
mean_gpa_tutoring = cleaned_data[cleaned_data['Tutoring'] == 1]['GPA'].mean()
mean_gpa_no_tutoring = cleaned_data[cleaned_data['Tutoring'] == 0]['GPA'].mean()

print(f"Середній GPA для учнів з додатковими заняттями: {mean_gpa_tutoring:.2f}")
print(f"Середній GPA для учнів без додаткових занять: {mean_gpa_no_tutoring:.2f}")

# Візуалізація розподілу GPA для учнів з додатковими заняттями та без них
plt.figure(figsize=(10, 6))
sns.boxplot(x='Tutoring', y='GPA', data=cleaned_data)
plt.title('Вплив додаткових занять на академічні успіхи')
plt.xlabel('Додаткові заняття (0 - Ні, 1 - Так)')
plt.ylabel('GPA')
plt.xticks(ticks=[0, 1], labels=['Без додаткових занять', 'З додатковими заняттями'])
plt.show()

# Виконання t-тесту для двох вибірок
t_stat, p_value = stats.ttest_ind(cleaned_data[cleaned_data['Tutoring'] == 1]['GPA'], cleaned_data[cleaned_data['Tutoring'] == 0]['GPA'])

print(f"t-статистика: {t_stat:.2f}")
print(f"p-значення: {p_value:.4f}")

if p_value < 0.05:
    print("Різниця в GPA між учнями з додатковими заняттями та без них є статистично значущою.")
else:
    print("Різниця в GPA між учнями з додатковими заняттями та без них не є статистично значущою.")
```

##### Гіпотеза Н0 про відсутність різниці в середніх значеннях GPA між учнями з додатковими заняттями та без них відхиляється. Результати показують, що додаткові заняття мають статистично значущий вплив на академічні успіхи учнів.

## 4. Аналіз участі батьків

#### Дослідити, як рівень підтримки батьків впливає на академічні результати учнів

```python
# Обчислення середніх значень GPA для кожного рівня підтримки батьків
mean_gpa_by_support = cleaned_data.groupby('ParentalSupport')['GPA'].mean().reset_index()
print(mean_gpa_by_support)

# Візуалізація розподілу GPA для різних рівнів підтримки батьків
plt.figure(figsize=(12, 6))
sns.boxplot(x='ParentalSupport', y='GPA', data=cleaned_data)
plt.title('Вплив рівня підтримки батьків на академічні успіхи')
plt.xlabel('Рівень підтримки батьків')
plt.ylabel('GPA')
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['None', 'Low', 'Moderate', 'High', 'Very High'])
plt.show()

# Виконання ANOVA тесту
anova_result = f_oneway(cleaned_data[cleaned_data['ParentalSupport'] == 0]['GPA'],
                        cleaned_data[cleaned_data['ParentalSupport'] == 1]['GPA'],
                        cleaned_data[cleaned_data['ParentalSupport'] == 2]['GPA'],
                        cleaned_data[cleaned_data['ParentalSupport'] == 3]['GPA'],
                        cleaned_data[cleaned_data['ParentalSupport'] == 4]['GPA'])

print(f"F-статистика: {anova_result.statistic:.2f}")
print(f"p-значення: {anova_result.pvalue:.4f}")

if anova_result.pvalue < 0.05:
    print("Різниця в GPA між групами з різним рівнем підтримки батьків є статистично значущою.")
else:
    print("Різниця в GPA між групами з різним рівнем підтримки батьків не є статистично значущою.")
```

##### Гіпотеза Н0 про відсутність різниці в середніх значеннях GPA між групами з різним рівнем підтримки батьків відхиляється. Результати показують, що рівень підтримки батьків має статистично значущий вплив на академічні успіхи учнів

### Визначити основні фактори, що впливають на класифікацію оцінок (GradeClass).

```python
# Переведення категоріальної змінної GradeClass у числовий формат, якщо вона не є числовою
data['GradeClass'] = cleaned_data['GradeClass'].astype(int)

# Огляд кореляцій між числовими змінними і GradeClass
correlation_matrix = cleaned_data.corr()
print("Кореляційна матриця:")
print(correlation_matrix['GradeClass'])

# Візуалізація кореляційної матриці
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix[['GradeClass']], annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Кореляція між числовими змінними і GradeClass')
plt.show()
```
